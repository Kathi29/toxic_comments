```python
#Drittes Dataset
#Toxic Russian Comments
#Link: https://www.kaggle.com/datasets/alexandersemiletov/toxic-russian-comments
data_list = []
with open("/Users/ekaterinakastaleva/dataset.txt") as file:
    for line in file:
        labels = line.split()[0]
        text = line[len(labels)+1:].strip()
        labels = labels.split(",")
        mask = [1 if "__label__NORMAL" in labels else 0,
                1 if "__label__INSULT" in labels else 0,
                1 if "__label__THREAT" in labels else 0,
                1 if "__label__OBSCENITY" in labels else 0]
        data_list.append((text, *mask))
```

```python
toxic_russian_comments = pd.DataFrame(data_list, columns=["text", "normal", "insult", "threat", "obscenity"])
```

```python
#To show the full text in the data
pd.set_option("display.max_colwidth", None)
```
